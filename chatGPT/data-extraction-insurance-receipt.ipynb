```python
import json
# https://uat.doc.com.ai/#/business/qhQeD9VHSL/project/v76sGkHu57zD1O/use
with open("./data/insurance_receipt_sample.json", "r") as file:
    data = json.load(file)
```

```python
output_data_full = []

output_data_text = []

output_data_string = ""

index = 0
for block in data['Blocks']:
    if block['BlockType'] == 'WORD':
        output_data_full.append({
            "text": block['Text'],
            "id": block['Id']
            #"id": index
        })
        output_data_text.append({
            "text": block['Text']
        })
        
        output_data_string += block['Text'] + ' '

        index += 1

output_json_full = json.dumps(output_data_full)
output_json = json.dumps(output_data_text)
print(output_data_full)
```
